```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load the dataset
file_path = '/kaggle/input/mall-customers/Mall_Customers.csv'  # Adjust this path if necessary
data = pd.read_csv(file_path)

# Select the features for clustering
features = data[['Annual Income (k$)', 'Spending Score (1-100)']]

# Create the K-means model with 3 clusters (for simplicity)
kmeans = KMeans(n_clusters=3, random_state=0)r

# Fit the model to the data
kmeans.fit(features)

# Get the cluster labels
labels = kmeans.labels_

# Add the cluster labels to the original dataset
data['Cluster'] = labels

# Plot the clusters
plt.scatter(data['Annual Income (k$)'], data['Spending Score (1-100)'], c=labels, cmap='viridis')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('Customer Segments')
plt.show()
```
